```python
import pandas as pd
```

```python
colunas = [
    'name',
    'situation',
    'neighborhood',
    'address',
    'number',
    'zip_code',
    'city',
    'state',
    'cnpj',
    'status',
    'additional_address_details',
    'main_activity',
    'latitude',
    'longitude',
]

empresas = pd.read_csv('2016-09-03-companies.xz', usecols=colunas)
print(empresas.shape)
empresas.head(3)
```

```python
empresas.columns
```

```python
empresas = empresas.query('status == "OK"')
empresas = empresas.query('situation == "ATIVA"')
print(empresas.shape)
empresas.head(3)
```

```python
empresas.to_csv(
    'empresas.xz',
    encoding='utf-8',
    index=False,
    compression='xz'
)
```
